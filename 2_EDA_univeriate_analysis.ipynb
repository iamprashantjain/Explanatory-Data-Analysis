```python
# uni: single
# veriate: variable
    
# uni-veriate means analyzing single column

# bi-variate: means analyzing 2 columns together
    
# multi variate means analyzing mutliple columns together
```

```python
# datatype

#data are of 2 types:

#Numerical data: means numbers like age, phone price, weight, height etc
#Categorical data: means data which can be categorized like country you belong, gender, religion you follow etc

#whenever you will have a data, it will have columns which will either be numerical or categorical
#univariate analysis means to find whether data is numerical or categorical?
```

```python
import pandas as pd

df = pd.read_csv('train.csv')
df.head()
```

```python
# uni-variate analysis: analyzing each column seperatly
#whether columns are numerical or categorical
```

# ** categorical data **

```python
df.columns
```

```python
import seaborn as sns
```

# countplot

```python
#how many people survived & died

sns.countplot(df['Survived'])
```

```python
#countplot is same as valuecount in pandas
#it will tell you the sum of categories in data
```

```python
#how many people were travelling in each pclass

sns.countplot(df['Pclass'])
```

```python
#how many pax boarded from which station

sns.countplot(df['Embarked'])
```

```python
sns.countplot(df['Sex'])
```

# pie chart

```python
#to get countplot info in %age
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
sns.distplot(df['Age'],bins=10)
```

```python
#pdf: probability density function
#if you ask what is the probability of age 40?
```

```python
#catch the line in y axis to xaxis
#0.15
```

```python
#histogram tells how many in each group
#displot tell pdf of any give number

#distplot also tells you about skeewness
#whether data is right skweed or left skweed
#it should be normally distributed
```
